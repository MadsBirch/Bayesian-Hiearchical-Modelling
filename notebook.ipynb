## Bayesian Hierarchical Modelling of Historical Co2 Emissions

In this project, we set out to model historical Green House Gas emmission from 1990 to 2018 using a Bayesian Hierarchical model. 

```python
import numpy as np
import pandas as pd

data = pd.read_csv('historical_emissions.csv', index_col=0)
data = data.drop(['World', 'European Union (27)'])

# we minus 2 becuase 2 of the entries are for the EU and World as wholes.
print(f'We have data from {data.shape[0]-2} countries.')
print(f'We have data from {data.shape[1]} years, from 1990 to 2018.')

data.head(5)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (18,12))
for country in data.index.values.tolist():
    data.loc[country][4:][::-1].plot(figsize =(20,12))

plt.title('Co2 Emission for all Countries from 1990 to 2018')
plt.show()
```
